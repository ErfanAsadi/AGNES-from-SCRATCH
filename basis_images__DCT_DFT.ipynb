# Extracting DFT and DCT 8 * 8 basis images
Author: Erfan Asadi

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import dct
```

### Making an image to calculate the basis images

```python
# defining the shape of the image
x_shape = 8 
y_shape= 8

np.random.seed(100)
image = np.eye(8)
```

### Calculating the fourier transform of the image

```python
#calculating the fast fourier transform of the Image
fft_image = np.fft.fft(image)
```

### Extracting FFT basis images 

```python
fft_real_part_basis_images = []
fft_imaginary_part_basis_images = []

for i in range(y_shape):
  x = fft_image[:][i]
  for j in range(x_shape):
    y=fft_image[:][j]
    basis_image = np.multiply.outer(x,np.transpose(y))
    fft_real_part_basis_images.append(basis_image.real)
    fft_imaginary_part_basis_images.append(basis_image.imag)
```

### Plot FFT Basis Images 

```python
fig = plt.figure(figsize=(8,8))
fig.title = 'Basis Images (Real Part)'
for i in range(1, x_shape * y_shape + 1):
    img = fft_real_part_basis_images[i-1]
    fig.add_subplot(8,8, i)
    plt.imshow(img,cmap='Greys')
plt.show()

fig = plt.figure(figsize=(8,8))
for i in range(1, x_shape * y_shape + 1):
    img = fft_imaginary_part_basis_images[i-1]
    fig.add_subplot(8,8, i)
    plt.imshow(img,cmap='Greys')
plt.show()
```

### Calculating DCT of image

```python
dct_image = dct(image)
dct_basis_images = []
for i in range(y_shape):
  x = dct_image[:][i]
  for j in range(x_shape):
    y = dct_image[:][j]
    basis_image = np.multiply.outer(x,np.transpose(y))
    dct_basis_images.append(basis_image)
```

### ploting DFT basis images 

```python
fig=plt.figure(figsize=(8,8))
for i in range(1, 8*8+1 ):
    img = dct_basis_images[i-1]
    fig.add_subplot(8,8, i)
    plt.imshow(img,cmap='Greys')
plt.show()
```
